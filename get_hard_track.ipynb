```python
df_tracks_day = df_tracks[df_tracks["time_day"] == "2019-01-25"]
df_tracks_day["time"] = pd.to_datetime(df_tracks_day["time"])
df_tracks_day = df_tracks_day.sort_values(by=["time"])
df_tracks_day.head()
```

```python
lat_center =  np.mean([df_tracks_day["lat"].min(), df_tracks_day["lat"].max()])
long_center =  np.mean([df_tracks_day["lon"].min(), df_tracks_day["lon"].max()])
```

```python

locations = df_tracks_day.iloc[:,2:4]
locationlist = locations.values.tolist()
times = df_tracks_day.iloc[:,1]
times 
```

```python
map_track = folium.Map(
    location = [ lat_center, long_center]   ,
    zoom_start = 14
)

# for i in range(0,len(df[2])):
#     folium.Marker(locationlist[i], popup=df.iloc[i][1]).add_to(map_track)
    
folium.PolyLine(locationlist, color="red", weight=2.5, opacity=1).add_to(map_track)

plugins.MarkerCluster(locationlist, popups = times).add_to(map_track) 
map_track.save(outfile= "test.html")
map_track
```
